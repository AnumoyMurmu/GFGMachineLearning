```python
import pandas as pd
```

```python
df=pd.read_csv("Dealing with duplicate data 02.csv")
```

```python
df
```

```python
df.duplicated().sum()
```

```python
df.duplicated()
```

```python
df[df.duplicated()]
```

```python
duplicates = df.duplicated(keep='first')
print(df[duplicates])  # This shows the rows with duplicates (excluding the first occurrence)
```

```python
duplicates = df.duplicated(keep='last')
print(df[duplicates])  #keeps the last duplicated values and shows first duplicated values
```

```python
duplicates = df.duplicated(keep=False)
print(df[duplicates]) 
```

```python
df.drop_duplicates()
```

```python
df.duplicated(["Name","City"]).sum()
```

```python
df.duplicated("City").sum()
```

### Example 2:

```python
data = {'CustomerID': [1001, 1002, 1003, 1004, 1005, 1006, 1007, 1008],
        'Name': ['Alice', 'Alice', 'Bob', 'Charlie', 'Charlie', 'David', 'Bob', 'Emily'],
        'Age': [25, 25, 30, 22, 22, 35, 30, 28],
        'City': ['New York', 'Miami', 'Los Angeles', 'Chicago', 'Miami', 'London', 'Los Angeles', 'Paris']}

df1 = pd.DataFrame(data)
```

```python
df1
```

```python
df1.duplicated().sum()
```

```python
df1.duplicated("City").sum()
```

```python
df1.drop_duplicates("City", keep="last")
```

```python
df1.drop_duplicates("City", keep="first")
```

```python
df1.drop_duplicates("City", keep=False)
```
