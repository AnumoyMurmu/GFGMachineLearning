```python
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
df.head(10)
```

```python
sns.countplot(x='survived', data=df, hue='class')
```

### Heatmap:
used when there are more categories

```python
x=pd.crosstab(df['survived'], df['class'])
```

```python
x
```

```python
sns.heatmap(x, annot=True, fmt='.1f', cmap='coolwarm')  #annot gives the values
```

```python
sns.barplot(x='class', y='age', hue='sex', data=df)
```

```python
sns.stripplot(x='class', y='age', hue='sex', data=df)
```
