```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt   ###added later
```

In univariate in study a particular column or field.

```python
df=pd.read_csv('Univariate Analysis non graphical.csv')
```

```python
df.head(10)
```

### Categorical:

#### Bar Graph:

```python
sns.countplot(data=df, x='neighbourhood_group')  ###data, x value, y value and hue as per documentation
```

```python
sns.countplot(data=df, x='room_type')
```

```python
### checking room distribution in a particular city

sns.countplot(data=df, x='neighbourhood_group', hue='room_type')
```

#### Pie Chart:

```python
df['room_type'].value_counts()
```

```python
wedgeprops = {'linewidth': 1, 'edgecolor': 'black'}

df['room_type'].value_counts().plot(kind='pie', autopct='%.2f', wedgeprops=wedgeprops)
```

```python
### For neighbourhood group

explode=[0, 0, .2, .1, .3]

wedgeprops = {'linewidth': 1, 'edgecolor': 'black'}
fig=plt.figure(figsize=(7,7))
df['neighbourhood_group'].value_counts().plot(kind='pie', autopct='%.2f', explode=explode, wedgeprops=wedgeprops, shadow=True)
```

```python

df['room_type'].value_counts().plot(kind='bar')
```
