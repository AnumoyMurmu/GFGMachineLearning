##### Bivariate analysis is a statistical method used to explore the relationship between two variables. It's a foundational step in data analysis, helping you identify potential patterns and associations between pairs of features within your data.

#### for documentation `shift+tab`

```python
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
df.head(10)
```

### Bar Graph:

```python
#age and class 
sns.barplot(x='class',y='age',data=df)
#showing average/mean (value) age of person travelling in different class
```

```python
sns.barplot(x='class', y='fare', data=df, estimator='sum')   #returns sum value
```

```python
sns.barplot(x='class',y='age',hue='sex', data=df)   #returns mean/average value
```

```python
sns.barplot(x='class',y='fare',hue='sex', data=df, estimator='sum') 
```

```python
sns.boxplot(x='class',y='age',data=df)
```

```python
sns.boxplot(x='class',y='age',hue='survived',data=df)
```

```python
sns.violinplot(x='class',y='age',hue='survived',data=df)
```

```python
sns.distplot(df[df['class']=='second']['age'])
```
