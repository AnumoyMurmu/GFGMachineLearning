```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df=sns.load_dataset('titanic')
df.head(10)
```

### Scatter Plot:

helps us find out the relationship between two variables.

```python
sns.scatterplot(x='age', y='fare', hue='sex', data=df)
```

```python
df2=sns.load_dataset('tips')
df2.head(10)
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=df2)
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', style='smoker', data=df2)
```

### JointPlot:

```python
sns.jointplot(x='total_bill', y='tip', data=df2)
```

```python
sns.jointplot(x='total_bill', y='tip', hue='sex', data=df2)
```

```python
sns.jointplot(x='total_bill', y='tip', data=df2, kind='reg')
```

```python
df2.corr()
```
