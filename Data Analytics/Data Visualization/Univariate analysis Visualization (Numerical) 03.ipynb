```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('Univariate Analysis non graphical.csv')
```

```python
df.head(2)
```

### Histogram:

```python
sns.histplot(data=df, x='price')
```

```python
sns.histplot(data=df[df['price']<500], x='price')
```

```python
sns.histplot(data=df, x='availability_365')
```

```python
#distplot gives probability distribution

sns.distplot(df['price'])
```

```python
sns.distplot(df['availability_365'])   #the blue line is probaility distribution
```

### Boxplot:

```python
sns.boxplot(df['price'])
```

```python
sns.boxplot(df['availability_365'])
```
