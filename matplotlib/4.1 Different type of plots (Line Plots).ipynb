```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt

%matplotlib inline
```

##### 4.1 Line plot

```python

x=np.linspace(0, 4*np.pi, 100)
y=np.sin(x)


plt.plot(x, y)


plt.title('x-y wave')

plt.show()
```

##### 4.1.2 Multiple line plots

```python
x=np.linspace(0, 4*np.pi, 100)

y_sin=np.sin(x)
y_cos=np.cos(x)

plt.figure(figsize=(15,8))
plt.style.use('Solarize_Light2')

plt.plot(x,y_sin, color='red', linewidth=5,label='sin')
plt.plot(x,y_cos, color='green', linewidth=5,label='cos')

plt.title('sin-cosine wave')

plt.legend(loc='upper right')

plt.show()
```

##### 4.1.4 We can also use linestyle commanf. it has 4 main styles solid, dashed, dashdot and dotted

```python
plt.figure(figsize=(15,5))
plt.plot(x,x+0, linestyle='solid', label='solid, (x,x+0)')
plt.plot(x,x+1, linestyle='dashed',label='dashed, (x,x+1)')
plt.plot(x,x+2, linestyle='dashdot',label='dashdot, (x,x+2)')
plt.plot(x,x+3, linestyle='dotted',label='dotted, (x,x+3)')

plt.legend(loc='upper right')
plt.show()
```

```python
x=np.linspace(0,100,101,dtype='int')
plt.figure(figsize=(15,5))
plt.plot(x,x+0, linestyle='solid', label='solid, (x,x+0)')
plt.plot(x,x+1, linestyle='dashed',label='dashed, (x,x+1)')
plt.plot(x,x+2, linestyle='dashdot',label='dashdot, (x,x+2)')
plt.plot(x,x+3, linestyle='dotted',label='dotted, (x,x+3)')

plt.legend(loc='upper right')
plt.show()
```

##### 4.1.5 Adjusting the plot axes
 1. can be done using plt.xlim for x axis
 2. plt.ylim for y axis

```python
x=np.linspace(0, 4*np.pi, 100)

plt.xlim([1,12])
plt.ylim([-2,2])

sin_y=np.sin(x)
plt.plot(x,sin_y)
plt.title('sine wave')
plt.axis('tight')
plt.show()
```

##### to avoid adjust grapg we can use plt.axis('tight') ot plt.axis('equal')

##### 4.1.6 using .set to combine all of these commands (everything is defined in one go)

```python
x=np.linspace(0, 4*np.pi, 100)

fig=plt.axes()
fig.plot(x,np.sin(x))

fig.set(xlim=(1,12), ylim=(-2,2),xlabel='x',ylabel='sin(x)',title='sin wave')
fig.plot()

```
