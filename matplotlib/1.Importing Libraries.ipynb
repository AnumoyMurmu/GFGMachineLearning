```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
np.linspace(0, np.pi, 10)
#starts from (zero, till pi value, has 10 intervals)
```

### Sine wave:

```python
x=np.linspace(0, np.pi, 10)
sin_y=np.sin(x)

print(x)
print(sin_y)
```

```python
plt.plot(x, sin_y)
plt.title('sine wave')
plt.show()
```

##### Multiplying the pi:

```python
x=np.linspace(0, 4*np.pi, 100)
sin_y=np.sin(x)

print(x)
print(sin_y)


plt.plot(x, sin_y)
plt.title('sine wave')
plt.show()
```

### Cos wave:

```python
x=np.linspace(0, 4*np.pi, 100)

#defining the wave
cos_y=np.cos(x)

# print(x)
# print(cos_y)

#plotting the wave
plt.plot(x, cos_y)

plt.title('cos wave')

plt.show()
```

### Sine and cos wave together:

```python
x=np.linspace(0, 4*np.pi, 100)

#defining the wave
sin_y=np.sin(x)
cos_y=np.cos(x)


# print(x)
# print(cos_y)

#plotting the wave
plt.plot(x, sin_y, label='sine')
plt.plot(x, cos_y, label='cosine')

plt.legend(loc='upper right')
plt.title('sine-cos wave')

plt.show()
```

```python
x=np.linspace(0, 4*np.pi, 100)

#defining the wave
sin_y=np.sin(x)
cos_y=np.cos(x)


# print(x)
# print(cos_y)

#plotting the wave
plt.plot(x, sin_y, '+')
plt.plot(x, cos_y, '-')

plt.title('sine-cos wave')

plt.show()
```
