```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt

%matplotlib inline
```

1. Used to visualize relationships between two variables where each data point represents a value pair.
2. Points are scattered on the x and y axes based on their corresponding values.
3. Created using the scatter function.

```python
n=10000
x=np.random.randn(n)   ##genearting random numbers
y=np.random.randn(n)   ##genearting random numbers

plt.scatter(x, y)
plt.title('Scatter plot')

plt.show()
```

```python
n=1000
x=np.random.randn(n)   ##genearting random numbers
y=np.random.randn(n)   ##genearting random numbers

plt.scatter(x, y,c='green',alpha=.6, s=40)
plt.title('Scatter plot')

plt.show()
```

alpha=for better clarirty/transparency of dots, 
s=size od dots

##### 4.2.2 Multiple Scatter plot

```python
n=1000
x1=np.random.randn(n)   ##genearting random numbers
y1=np.random.randn(n)   ##genearting random numbers

x2=np.random.randn(n) + 2   ##genearting random numbers
y2=np.random.randn(n) + 2

plt.scatter(x1,y1,c='green',alpha=.7, s=10, label='group-1')
plt.scatter(x2,y2,c='blue',alpha=.7, s=10, label='group-2')

plt.title('Scatter plot with multiple groups')

plt.legend(loc='upper right')

plt.show()
```

##### Colormap and Color Values:

The cmap='virdis' argument defines a colormap for the scatter points based on the values in the colors array. However, the colors array is generated using rng.rand(100), which creates random values between 0 and 1 (inclusive). These values might not map well to the chosen colormap 'virdis'.

To ensure a wider range of colors and a smoother gradient, you could consider scaling the colors array or using a different colormap like 'viridis' (without the 'd').

##### Scatter Point Size Range:

The sizes array is generated using rng.rand(100) * 1000. This creates random values between 0 and 1000 (exclusive).

Depending on your plotting window size, some points might appear very large or very small. You might want to adjust the scaling factor (e.g., multiplying by a smaller value like 100) to control the range of point sizes.



```python
rng = np.random.RandomState(0)  # For reproducibility

x = rng.randn(100)
y = rng.randn(100)

# Scale colors for better mapping to the colormap
colors = rng.rand(100) * 0.8  # Values between 0 and 0.8

# Adjust size scaling factor for better visualization
sizes = rng.rand(100) * 200  # Values between 0 and 200

plt.scatter(x, y, c=colors, s=sizes, alpha=1, cmap='viridis', edgecolors='black')
plt.title('Random Scatter plot')
plt.xlabel('x values')
plt.ylabel('y values')
plt.colorbar()
plt.show()
```
