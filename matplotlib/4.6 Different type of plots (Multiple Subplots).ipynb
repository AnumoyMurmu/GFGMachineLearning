```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt

%matplotlib inline
```

Matplotlib's subplots functionality allows you to create multiple panels within a single figure window. This is useful for visualizing multiple related plots or data comparisons on the same page. Here's a breakdown of the concept:

##### Creating Subplots:

The most common way to create subplots is using the plt.subplots function. It takes arguments to define the layout (rows and columns) of the subplots:

plt.subplots(nrows=m, ncols=n): Creates a grid of m rows and n columns of subplots.

`Index based`

```python
x=np.linspace(0,10,50)

fig, ax=plt.subplots(2,2)
                    #h and w
y=np.sin(x)
ax[0,0].plot(x,y)
ax[0,0].set_title('sin wave')

y=np.cos(x)
ax[0,1].plot(x,y)
ax[0,1].set_title('cos wave')

y=np.tan(x)
ax[1,0].plot(x,y)
ax[1,0].set_title('tan wave')


y=np.random.randn(1000)
ax[1,1].hist(y,bins=40)
ax[1,1].set_title('Random distribution')


plt.tight_layout()
plt.show()
```

```python
plt.subplots_adjust(hspace=.4,wspace=.4)

for i in range(1,7):
    plt.subplot(2,3,i)
    plt.text(0.5,0.5,str((2,3,i)), fontsize=15,ha='center')
```
