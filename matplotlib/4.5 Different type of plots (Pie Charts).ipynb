```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt

%matplotlib inline
```

`wedgeprops`

```python
cars=['Audi','bmw','tesla','hyundai']
data=[45,12,10,100]

wedgeprops = {'linewidth': 1, 'edgecolor': 'black'} 

fig=plt.figure(figsize=(7,7))
plt.pie(data, labels=cars, wedgeprops=wedgeprops)

plt.show()
```

##### 4.5.2 Basic Operations

```python
cars=['Audi','bmw','tesla','hyundai']
data=[45,12,10,100]
explode=[.1, .2, .3, .1]

wedgeprops = {'linewidth': 1, 'edgecolor': 'black'}
fig=plt.figure(figsize=(7,7))
plt.pie(data, labels=cars, explode=explode, wedgeprops=wedgeprops)

plt.show()
```

```python
cars=['Audi','bmw','tesla','hyundai']
data=[45,12,10,100]
explode=[0, 0, .3, .1]

wedgeprops = {'linewidth': 1, 'edgecolor': 'black'}
fig=plt.figure(figsize=(7,7))
plt.pie(data, labels=cars, explode=explode, wedgeprops=wedgeprops)

plt.show()
```

`Adding Shadow`

```python
cars=['Audi','bmw','tesla','hyundai']
data=[45,12,10,100]
explode=[0, 0, .3, .1]

wedgeprops = {'linewidth': 1, 'edgecolor': 'black'}
fig=plt.figure(figsize=(7,7))
plt.pie(data, labels=cars, explode=explode, wedgeprops=wedgeprops, shadow=True)

plt.show()
```

`Adding Percentage`

```python
cars=['Audi','bmw','tesla','hyundai']
data=[45,12,10,100]
explode=[0, 0, .3, .1]

wedgeprops = {'linewidth': 1, 'edgecolor': 'black'}
fig=plt.figure(figsize=(7,7))
plt.pie(data, labels=cars, explode=explode, wedgeprops=wedgeprops, shadow=True, autopct='%1.1f%%')

plt.show()
```
