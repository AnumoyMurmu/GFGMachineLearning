### 11. to_csv() function:

```python
data = {
    'one': pd.Series([1, 2, 3, 4]),    # Series for 'one' column
    'two': pd.Series([5, 6, 7, 8]),    # Series for 'two' column
    'three': pd.Series([9, 10, 11]),   # Series for 'three' column (with fewer values)
    'four': pd.Series([12, 13, 14, 15]),# Series for 'four' column
    'five': pd.Series([16, 17, 18, 19]) # Series for 'five' column
}

df=pd.DataFrame(data)
```

```python
df.to_csv('Number.csv')   ##converts to csv file
```
